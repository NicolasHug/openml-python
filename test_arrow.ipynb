```python
list_datasets = datasets.list_datasets(output_format="dataframe")
list_datasets_feats = list_datasets.sort_values(by="NumberOfFeatures", ascending=False)[:100]
#list_datasets_ins = list_datasets.sort_values(by="NumberOfInstances", ascending=False)[1:5]
list_datasets = list_datasets_feats
```

```python
most_instances = list_datasets["did"].values
```

```python
writef = []
readf = []
size = []
for did in most_instances:
    print(did)
    data = datasets.get_dataset(int(did))
    x, y, categorical, attribute_names = data.get_data()
    df = pd.DataFrame(x, columns=attribute_names)
    start = time.time()
    try:
        feather.write_feather(df, 'example_feather')
    except:
        pass
    end = time.time()
    writef.append(end-start)
    start = time.time()
    try:
        feather.read_feather('example_feather')
    except:
        pass
    end = time.time()
    readf.append(end-start)
    size.append(df.memory_usage(index=True).sum() * 1e-9)
```

```python
writep = []
readp = []
for did in most_instances:
    print(did)
    data = datasets.get_dataset(int(did))
    x, y, categorical, attribute_names = data.get_data()
    df = pd.DataFrame(x, columns=attribute_names)
    start = time.time()
    df.to_pickle("example_pickle.pkl") 
    end = time.time()
    writep.append(end-start)
    start = time.time()
    pd.read_pickle("example_pickle.pkl")
    end = time.time()
    readp.append(end-start)
    
```

```python
list_datasets = list_datasets[["did", "name", "NumberOfFeatures", "NumberOfInstances"]]
list_datasets["featherwrite"] = writef
list_datasets["featherread"] = readf
list_datasets["picklewrite"] = writep
list_datasets["pickleread"] = readp
list_datasets["memoryusage(GB)"] = size
list_datasets
```
